# 07: DATOS FALTANTES

```python
import pandas as pd
import os
```

```python
# importamos datos

raiz = r"C:\Users\Usuario\Documents\05_Formación_Académica\Marc\Data_Science\machine_learning\Recursos_curso\datasets"
fichero = r"titanic\titanic3.csv"
ruta = os.path.join(raiz, fichero) 
data = pd.read_csv(ruta)
```

### Comprobar los valores que faltan en el dataframe

```python
# Función para visualizar los valores que son null (N/A) dentro del dataframe

pd.isnull(data) # sustituye todos los valores por un True (si es N/A) o un False (si NO es N/A)
```

```python
# Se puede visualizar una única columna del mismo modo

pd.isnull(data["body"])
```

```python
# Contar el número de datos nulos dentro del dataframe

pd.isnull(data).values # devuelve un array de que contiene tantas listas de valores True o False como columnas tiene el dataframe
pd.isnull(data).values.ravel() # devuelve una unica lista de valores True o False
pd.isnull(data).values.ravel().sum() # suma todos los valores True (que se interpretan como 1) y los false los obvia (se interpretan como 0)
```

```python
# Visualizar cuantos nulos hay por columnas

columnas = data.columns.values

for i in columnas:
    nulos = pd.isnull(data[i]).values.ravel().sum()
    print(f"Valores nulos en la columna {i}: {nulos}")
```

### Eliminar los valores nulos del dataframe

```python
# Eliminar los datos que tengan un valor nulo en TODAS las columnas

data.dropna(axis = 0, # si se especifica 0 borrará las FILAS que contengan nulos. SI se selecciona 1 borrara las COLUMNAS
           how = "all") # si seleccionamos el tipo 'all' solo se borraran las filas que TODAS las columnas sean nulas
```

```python
# Eliminar los datos que tengan CUALQUIER valor nulo en su fila

data.dropna(axis = 0,
           how = "any") # si seleccionamos el tipo 'any' se borraran las filas que tengan CUALQUIER valor nulo
```

```python
# Eliminar todas las columnas que tengan algun valor nulo

data.dropna(axis = 1, how = "any").head() # como se puede ver, nos hemos quedado con menos columnas que en el dataframe original, pero sin valores nulos
```

### Inferir datos nulos (reemplazar valores valtantes por otros como por ejemplo la media o un string)

```python
# Reemplazar los valores nulos por un valor como por ejemplo el 0

data.fillna(0).head()
```

```python
# Reemplazar los valores nulos por una string

data.fillna('Unknown').head()
```

```python
# Reemplazar los valores de una única columna por una string

data['body'] = data['body'].fillna("Unknown") # de esta forma estamos sobrescribiendo en el dataframe. Antes No estabamos sobreescribiendo, solo viendo
data.head()
```

```python
# Reemplazar valores de una única columna por la media de la columna

data['age'] = data['age'].fillna(int(data['age'].mean())) # ponemos la media de edad y para que no salgan decimales lo hacemos int()
data.head(50)
```

```python
# Remplazar un valor nulo por el valor NO nulo más proxima de la celda de arriba 

valor_original = data['boat'][2]
valor_nuevo = data['boat'].fillna(method = 'ffill') # esto esta en desuso, en el futuro no sera un metodo 'ffill' sino que sera funcion .ffill()

print(f'La celda boat 2 tiene una valor original de {valor_original} y se cambia por un {valor_nuevo[2]} que es el mismo valor de la celda boat 1')
```

```python
# Remplazar un valor nulo por el valor NO nulo más proxima de la celda de abajo

valor_original = data['cabin'][9]
valor_nuevo = data['cabin'].bfill() # aqui vemos como funciona la nueva forma de utilizar el reemplazo de un nulo por un valor proximo

print(f'La celda cabin 9 tiene una valor original de {valor_original} y se cambia por un {valor_nuevo[9]} que es el mismo valor que cabin 10')
```
