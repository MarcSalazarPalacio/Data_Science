```python
# Crear un subconjunto con todas las columnas exceto las de una lista de NO deseadas

columnas_no_deseadas = ['Day Calls', 'Night Calls', 'Phone', 'Account Length'] # lista de columnas no deseadas
todas_columnas = data.columns.values.tolist() # lista de todas las columnas del dataset original
```

```python
lista_buena = [i for i in todas_columnas if i not in columnas_no_deseadas] # forma de cruzar ambas listas y quedarnos solo con los unicos en la primera
lista_buena
```

```python
new_df = data[lista_buena]
new_df.head()
```

```python
# otra forma de cruzar ambas listas (AL HACERLO CON SETS, EL ORDEN SE PIERDE)

set_no_deseado = set(columnas_no_deseadas)
set_todo = set(todas_columnas)

lista_deseada = list(set_todo - set_no_deseado)
```

### Crear un subconjunto de FILAS

```python
# Escoger un conjunto de filas especificas

filas_0_5 = data[0:6] # el primer argumento incluye la fila y el segundo argumento NO, por lo que se queda con la fila anterior
filas_0_5
```

### Filtrar filas respecto a condiciones

```python
# Filtrar por una string de una celda

data[data['State'] == "CT"].head()
```

```python
# Filtrar por el valor de una celda

data[data['Day Calls'] > 100].head()
```

```python
# Filtrar en base a operaciones entre dos columnas

data[data['Day Calls'] > data['Night Calls']].head()
```

### Filtrar filas respecto a más de una condición

```python
# Que se cumplan AMBAS condiciones "AND --> &"

data[(data['Day Calls'] > 100) & (data['State'] == "NY")].head() # se ha de poner cada condicion entre ()
```

```python
# Que se cumpla CUALQUIER condiciones "OR --> |"

data[(data['Day Calls'] > 100) | (data['State'] == "NY")].head()
```

### Filtrar FILAS y COLUMNAS mediante el método data[][]

```python
# Quedarse solo con las columnas por las que hemos filtrado, es decir, filtrar FILAS y COLUMNAS

data[(data['Day Calls'] > 100) | (data['State'] == "NY")][['State', 'Day Calls']].head() # tenemos un formato tal que data[FILAS][COLUMNAS]
```

### Filtrar FILAS y COLUMNAS mediante el método ILOC[]

```python
# Siempre van primero las filas y luego las columnas

data.iloc[0:6, # FILAS
          1:4] # COLUMNAS
```

```python
# seleccionar filas o columnas no consecutivas

data.iloc[0:5, [1, 3, 5]]
```

```python
# Seleccionar todas las filas/columnas de unas determinadas columnas/filas

data.iloc[0:5, :] # todas las columnas de 5 filas
```

### FIltrar filas mediante el método loc[]

```python
# Sirve para llamar a las columnas por su nombre

data.loc[0:6, ['State', 'Area Code']]
```
