### Cargar el fichero con diferentes parametros ###

```python
# Cargar el fichero .csv mediante su path

data_frame = pd.read_csv(filepath_or_buffer = path, # especificar el directorio del dataframe
                        sep = ",", # especificar como se separa el documento
                        dtype = None, # especificar el tipo de dato de cada columna, por defecto es None
                        header = 0, # especificar que fila se usa como cabecera, por defecto es 0 (la primera fila)
                        names = None, # especificar los nombres de cada columna. Se le puiede pasar un array o lista con los nombres
                        skiprows = None, # especifica el numero de filas que NO vamos a tener en cuenta en el dataframe respecto a la priemra fila
                        index_col = None, # especificar el indice de las filas. Se puede hacer que una columna sea el indice de una fila
                        skip_blank_lines = False, # especificar si queremos saltarnos los valores nulos (NaN)
                        na_filter = False, # especificar si queremos borrar las filas con valores nulos (NaN)
                        )
```
