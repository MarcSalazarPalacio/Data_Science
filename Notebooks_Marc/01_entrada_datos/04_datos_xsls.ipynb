# 04: CARGAR FICHEROS XLSX #

```python
# Importar las librerias

import pandas as pd
import os
```

```python
# Establecer el fichero que vamos a leer

raiz = r"C:\Users\Usuario\Documents\05_Formación_Académica\Marc\Data_Science\machine_learning\Recursos_curso\datasets"
fichero = r"titanic\titanic3.xlsx"
ruta = os.path.join(raiz, fichero) 
```

### Leer un fichero excel ###

```python
# Leer el fichero

data_xlsx = pd.read_excel(ruta)
```

```python
data_xlsx.head()
```

```python
# Leer el fichero especificando la hoja que vamos a leer

data_hoja = pd.read_excel(ruta, 'Hoja_2')
```

### Crear un nuevo archivo a partir de este dataframe creado ###

```python
# Crear un excel nuevo

nuevo_fichero = r"titanic\ejemplo_excel.xlsx"

data_xlsx.to_excel(os.path.join(raiz, nuevo_fichero))
```

```python
# Crear un csv nuevo

nuevo_fichero = r"titanic\ejemplo_csv.csv"

data_xlsx.to_csv(os.path.join(raiz, nuevo_fichero))
```
