# 05: EJERCICIO A PARTIR DE UNA URL #

#### El siguiente ejercicio consiste en leer los datos que hay en una url mediante la libreria urlib3, procesarlos y obtener un dataframe ###

```python
# Cargar librerias

import pandas as pd
import urllib3
```

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRgDmJGDkg6EYP8aTOyOM2bU1Q8PBCi9HImsXr-MVKj08gJkG_c5OsfiOUCmYYLzUWa6-diTJXdu60K/pub?output=csv"
```

### Cargar los datos mediante la libreria urlib3 ###

```python
# Creamos un 'pool manager' que gestionar√° las peticiones

http = urllib3.PoolManager()
```

```python
# Obtenemos los datos de la url

data = http.request("GET", url)
```

```python
# Visualizamos los datos que hemos obtenido. En este caso, se trata de un dato binario

binary_data = data.data
binary_data
```

### Decodificar los datos binarios en UTF-8 y separamos el texto por delimitaciones###

```python
# Decodificamos los datos para tener una string

string_data = binary_data.decode('utf-8')
string_data
```

```python
# Separamos la string por delimitadores de salto de l√≠nea

lined_data = string_data.split('\n')
lined_data
```

### Preparamos los datos pora generar el diccionrio ###

```python
# Creamos la cabecera

cabecera = lined_data[0].strip().split(",")
cabecera
```

```python
# Creamos el diccionario y le ponemos la cabecera

diccionario = {}

for i in cabecera:
    diccionario[i] = []

diccionario
```

```python
# Rellenamos el diccionario con los datos de la string

lined_data.remove(lined_data[0]) # removemos la cabecera para que no se vuelva a poner

for i in lined_data:
    new_line = i.strip().split(",")
    for j in range(len(cabecera)):
        diccionario[cabecera[j]].append(new_line[j])
```

```python
diccionario
```

### Creamos el dataframe a partir del diccionario creado ###

```python
# Creamos y visualizamos el dataframe

dataframe = pd.DataFrame(diccionario)
dataframe.head()
```
