### Crear un dataframe a partir del documento ###

```python
# Volver a leer el documento para vovler al principio

data = open(ruta, "r") # si ponemos una r significa que solo vamos a leer el fichero
```

```python
# Guardar los nombres de cada columna en una lista

nombres = data.readline().strip().split(",") # strip() elimina los espacios del inicio y final, y split() especifica el separador que usamos
nombres
```

```python
# Crear un diccionario vacío y crear las diferentes keys a partir de los nombres de las columnas

diccionario = {} # crea un diccionario vacío

for i in nombres:
    diccionario[i] = [] # genera una key nueva con una lista vacía como values por cada uno de los nombres de las columnas

diccionario
```

```python
# Llenar cada una de las keys del diciconario con las diferentes filas del fichero

for i in data: # i va a representar cada una de las líneas del fichero sin contar el encabezado que ya lo hemos leido antes
    valores = i.strip().split(",") # de cada línea vamos a crear una lista con cada uno de los parametros de la fila
    for j in range(len(nombres)): 
        diccionario[nombres[j]].append(valores[j]) # metemos el valor adecuado en la llave correspondiente del diciconario
```

```python
# Crear el dataframe

import pandas as pd

data_frame = pd.DataFrame(diccionario)
```

### Escribir un nuevo fichero para cambiar el tipo de separador ###

```python
# Crear la nueva ruta del fichero que vamos a crear

fichero_nuevo = r"titanic\titanic3.txt"
ruta_nueva = os.path.join(raiz, fichero_nuevo)
```

```python
# Leer el fichero antiguo de nuevo

data = open(ruta, "r")
```

```python
# Crear un nuevo archivo txt vacío

nueva_data = open(ruta_nueva, "w") # con el modo "w" creamos un archivo vacío o borramos lo que hay en uno ya creado
```

```python
# Escribir el nuevo fichero a partir del antiguo pero cambiando la separación por una tabulación

for i in data:
    linea = i.strip().split(",") # crea una variable en formato lista con cada uno de los valores de u a fila
    nueva_data.write("\t".join(linea)) # junta TODOS los valores de la lista creada antes a traves de un "\t" y lo escribe en el nuevo fichero
    nueva_data.write("\n") # escribe un salto de línea
```

```python
# Cragar el nuevo fichero que esta separado por \t

nuevo_dataframe = pd.read_csv(ruta_nueva,
                              sep = "\t")
nuevo_dataframe.head()
```
